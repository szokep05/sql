## Termékkategóriák száma

```
SELECT COUNT(*) FROM Termekkategoria;
```

Az adatbázis **155 különböző termékkategóriát** tartalmaz. Ez azt mutatja, hogy a kínálat széles skálán mozog.

## Egyedi árakhoz tartozó termékek száma

```

SELECT LISTAAR, COUNT(*) AS 'Egyedi'
FROM Termek
GROUP BY LISTAAR
HAVING COUNT(*) = 1
ORDER BY LISTAAR DESC;
```

Az adatok alapján **115 különböző ár** csak egy-egy termékhez tartozik. Ez azt jelzi, hogy sok termék egyedi árazású, nem tömegesen használt árral rendelkezik.

## Duplikált árakhoz tartozó termékek száma

```

SELECT LISTAAR, COUNT(*) AS 'duplikáció'
FROM Termek 
GROUP BY LISTAAR 
HAVING COUNT(*) > 1;
```

Az adatok szerint **105 árhoz több termék is tartozik**, tehát vannak ismétlődő árkategóriák. Ez arra utalhat, hogy az árazás szabványosított bizonyos termékek esetében.

## Minimális, maximális és átlagos rendelt mennyiség

```

SELECT 
    MIN(MENNYISEG) AS 'Min',
    MAX(MENNYISEG) AS 'Max',
    AVG(MENNYISEG) AS 'Átlag'
FROM Rendeles_tetel;
```

Az adatok alapján:
- **Legkisebb rendelt mennyiség**: 1
- **Legnagyobb rendelt mennyiség**: 90
- **Átlagos rendelési mennyiség**: ~39,2
Ez arra utal, hogy a rendelések többsége kisebb mennyiségekre koncentrálódik, de előfordulnak nagyobb tételek is.

## Legelső és legutolsó rendelés dátuma

```

SELECT rend_datum
FROM Rendeles
WHERE rend_datum = (SELECT MIN(rend_datum) FROM Rendeles);
```

```

SELECT rend_datum
FROM Rendeles
WHERE rend_datum = (SELECT MAX(rend_datum) FROM Rendeles);
```

Az első rögzített rendelés **2015. május 1-jén** történt, míg az utolsó **2017. április 1-jén**. Ez azt jelenti, hogy az adatbázis kb. **két évnyi rendelési adatot** tartalmaz.
